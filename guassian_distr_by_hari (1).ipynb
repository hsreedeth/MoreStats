## Importing Necessary Libraries 
- numerical py or numpy is an essential library used as a tool in statistical analysis and advanced mathematical operations using arrays. 
- matplotlib is a powerful library capabale of interactive statistical visualizations.
- scipy.stats is a module that contains a large number of probability distributions, summary and frequency statistics, correlation functions and statistical tests, masked statistics, kernel density estimation, quasi-Monte Carlo functionality, and more.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

- The following steps demonstrate creating an np array that has our sample data. Next, we derive the mean and standard deviation values from the array data.
- The dataset used for this experiment is as following:

> **_15,12,8,7,16,13_**

```python
data = np.array([15,12,8,7,16,13])

mean = np.mean(data)
std_dev = np.std(data)
```

```python
mean
```

```python
std_dev
```

- The next step demonstrates setting up the x and y axis for the plot that will visualize the normal distribution. 
- mean - 3*std_dev and mean + 3*std_dev define the range on the x-axis that covers approximately three standard deviations on each side of the mean (based on the 68-95-99.7% rule). This ensures that you capture the majority of the data points and the shape of the distribution curve.
- The y-axis is made using the pdf function of the scipy.stats module. The the overall shape of the bell curve in a normal distribution is described by the Probability Density Function (PDF). 

```python
x = np.linspace(mean - 3*std_dev, mean + 3*std_dev, 100)
y = norm.pdf(x, mean, std_dev)
```

```python
plt.plot(x, y, color='red', alpha=0.75)
plt.xlabel('Range')
plt.ylabel('Probability Density')
plt.title('Gaussian Distribution')
plt.show()
```

- Let's try to find the PDF of a given number based on the calculations and dataset that we have. Let's assume that we would like to find the PDF at the value 7.
- Here is how to do it

- Note that this is only a Probability Density Function and does not resemble the probability.


```python
pdf_at_7 = norm.pdf(7, mean, std_dev)
pdf_at_7
```

- Probability Interval for the range 6.5 and 7.5

```python
prob_intvl = norm.cdf(7.5, mean, std_dev) - norm.cdf(6.5, mean, std_dev)
prob_intvl
```

## Z-Score Transformation (Standardization)

- Standardization (also known as z-score transformation) is a technique used to convert individual data points from any normal distribution into a standard normal distribution (also called the Z-distribution). 
- The standard normal distribution has a mean of 0 and a standard deviation of 1. This process allows you to compare data from different normal distributions on a common scale.

```python
ZScore = (data - mean) / std_dev
```

```python
ZScore
```

```python
data
```

```python
fin = dict(zip(data, ZScore))
fin
```

- When you standardize a data point using this formula, the resulting z-score tells you how many standard deviations the data point is away from the mean.
- If the z-score is positive, the data point is above the mean; 

> **_As observed, 16 has a Z-Score of 1.24 indicating that it is well above the Mean_**

- if it's negative, the data point is below the mean.

> **_7, on the other hand has a Z-Score of -1.44 indicating that it is below the Mean_**
